# Questão 2) B) Diagrama de Polos e Zeros

# Questão 2) C) Diagrama de Bode

Farei uso da "TransferFunction" e a "bode" da biblioteca Signal. A primeira retorna a Função de Transferência associada aos parâmetros da entrada e a segunda retorna a frequência, a magnitude e a fase dada uma array de intervalos de tempo. Usarei funções da "matplot" para plotar o Diagrama de Bode, que possui escala logarítmica.

```
sys = signal.TransferFunction([6,6],[1,-15])
w = np.logspace(-2,5,num=100, base=10.0)
w, mag, phase = signal.bode(sys,w=w,n=200)
```

Magnitude

```
plt.figure(figsize=(8,6))
plt.semilogx(w,mag)
plt.grid(True, which="both", ls="-")
plt.xlabel('Omega (rad/s)');
plt.ylabel('Mag (dB)');
```

Fase

```
plt.figure(figsize=(8,6))
plt.semilogx(w,phase)
plt.grid(True, which="both", ls="-")
plt.xlabel('Omega (rad/s)');
plt.ylabel('Phase (rad)');
```

#Questão 2) F) Resposta ao Degrau Unitário

Farei uso da "TransferFunction" e "step" da biblioteca Signal. A primeira retorna a Função de Transferência associada aos parâmetros da entrada e a segunda retorna valores de tempo e a resposta do sistema ao Degrau Unitário neles. Usarei funções da "matplot" para plotar a resposta "yout" pelo tempo.



```
sys = signal.TransferFunction([6,6],[1,-15])
tout, yout = signal.step(sys)
plt.plot(tout, yout, 'k', linewidth=1.5)
plt.grid(True, which="both", ls="-")
plt.xlabel('t');
plt.ylabel('Y out(t)')
```

# Questão 2) G) Resposta a Rampa Unitária

Farei uso da "TransferFunction" e "lsim" da biblioteca Signal. A primeira retorna a Função de Transferência associada aos parâmetros da entrada e a segunda retorna valores de tempo e a resposta do sistema neles. A função da Rampa Unitária será a prórpia função do tempo criada pela "linspace". Usarei funções da "matplot" para plotar a resposta "yout" pelo tempo.

```
sys = signal.TransferFunction([6,6],[1,-15])
time = np.linspace(0, 100, 1000)
tout, yout, xout = signal.lsim(sys, time, time)

plt.plot(tout, yout, 'k', linewidth=1.5)
plt.grid(True, which="both", ls="-")
plt.xlabel('t');
plt.ylabel('Y out(t)')
```

# Questão 2) H) Resposta a uma onda quadrada de frequência (π/4)

Farei uso da "TransferFunction", "square" e "lsim" da biblioteca Signal. A primeira retorna a Função de Transferência associada aos parâmetros da entrada, a segunda gera uma onda quadrada dada uma frequência e a terceira retorna valores de tempo e a resposta do sistema neles. Usarei funções da "matplot" para plotar a resposta "yout" pelo tempo.

```
time = np.linspace(0, 100, 1000)
sys = signal.TransferFunction([6,6],[1,-15])
entrada = signal.square((np.pi/4)*time)
tout, yout, xout = signal.lsim(sys,entrada,time)
plt.plot(tout, yout, 'k', linewidth=1.5)
plt.grid(True, which="both", ls="-")
plt.xlabel('t');
plt.ylabel('Y out(t)')
```
